### Spanish population growth in the 21st century

```python
europe_21st = europe_pop[list(np.arange(2000,2021))].T
europe_21st.index.name = 'Year'
europe_21st
```

```python
def growth(country_pop):
    initial_pop = country_pop.iloc[0]
    final_pop = country_pop.iloc[-1]
    return (final_pop - initial_pop) / initial_pop * 100
print('European countries population growth along the 21st century:')
pop_growth_21st_cent = europe_21st.apply(growth).sort_values()
pop_growth_21st_cent = pop_growth_21st_cent.apply(lambda x: f'{x:.2f} %')
pop_growth_21st_cent.name = '21st cent pop growth'
pop_growth_21st_cent.index.name = 'Country'
pop_growth_21st_cent
```
